```python
#Primera prueba de hipotesis

# Una fabrica 135 unidades por hora con una desviacion estandar  de 8 unidades por hora, una maquina nueva produce  160 unidades/hora con una desviacion estandar de 8
# total de muestras 35
# el nivel de significancia lo establecemos alfa = 0.01 porque la maquina es cara y queremos ser cautos


z_score = (160 - 150)/ (8/np.sqrt(35))
print(z_score)
p_value = 2*(1-stats.norm.cdf(abs(z_score)))
print(p_value)
```
